```python
def custom_loss_func(y_pred , y_true):
    return (y_pred - y_true)**2
```

```python
model_1 = tensorflow.keras.models.Sequential([
    tensorflow.keras.layers.Dense(31),
    tensorflow.keras.layers.Dense(100),
    tensorflow.keras.layers.Dense(100),
    tensorflow.keras.layers.Dense(2)
])
# ypred = model
# y_true =
model_1.compile(optimizer = tensorflow.keras.optimiers.Adam() , loss = custom_loss_func , metric = ['accuracy'])
```
